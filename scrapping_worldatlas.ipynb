```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
import re
import pandas as pd

html = urlopen("http://www.worldatlas.com/cntycont.htm") #This is link
soup = BeautifulSoup(html, "lxml") #Parsing using lxml

con_list = [] #Country List
cot_list = [] #Region List

'''
The page is somewhat simple doesn't have class for list, so I struggled to grab list of countries 
Initially. But, I found 'H2' heading of region and then countries in his next to next sibling. 
I used this to extract info
'''

for ele in soup.find_all('h2',limit = 6): 
    d = ele.next_sibling.next_sibling.get_text() #getting next to next sibling
    con_list.append(d.split('\n')) #I got whole country list in string
    cot_list.append(ele.get_text()) #Also I got region list

e = [] # Creating empty list for storing Country list
c = [] # Creating empty list for storing Region list
i = 0 # iterating 

for ele in con_list: #con_list is big string of 6
    for con in ele: #iterating over one big string
        if(con != ''): #There were some empty element
            e.append(con) #extracting each row from string
            da = cot_list[i] #test
            da = re.sub('[{\n}(0-9)]','',da) #using regex to string data .. (Region were in AFRICA(46)
                                            # so, I used regex to replace \n and (*))
            c.append(da)
            
    i = i + 1 #Increasing the iterate

data = pd.DataFrame(e)
data.columns = ['Countries']
data['Region'] = c
print(data)
```
